# Read PDF Content With PyPDF2

```python
import PyPDF2
```

```python
pdf_file = open("Samsung.pdf", 'rb')
```

```python
pdf_reader = PyPDF2.PdfReader(pdf_file)
```

```python
total_pages = len(pdf_reader.pages)
file_content = ""
for i in range(total_pages):
    if(i == 0):
        file_content = pdf_reader.pages[0].extract_text()
    else:
        file_content += f"\n\n<==========>\n\n{pdf_reader.pages[i].extract_text()}"
with open('extracted_text.txt', 'w+') as new_file:
    new_file.write(file_content)
```
